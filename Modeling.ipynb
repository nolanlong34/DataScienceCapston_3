```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train = np.array(pd.read_csv('X_train.csv'))
X_test = np.array(pd.read_csv('X_test.csv'))
Y_train = np.array(pd.read_csv('Y_train.csv'))
Y_test = np.array(pd.read_csv('Y_test.csv'))
```

```python
# Create a logistic regression model
model = LogisticRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
cm = confusion_matrix(Y_test, Y_pred)
print(cm)

```

```python
# Create a decision tree model
param_grid = {'max_depth': np.arange(3, 10), 'criterion': ['gini', 'entropy'], 'splitter': ['best', 'random'], 'min_samples_split': np.arange(2, 10), 'min_samples_leaf': np.arange(1, 10)}
model = tree.DecisionTreeClassifier()
grid = GridSearchCV(model, param_grid, scoring='accuracy')
grid.fit(X_train, Y_train)
Y_pred = grid.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
print(confusion_matrix(Y_test, Y_pred))
print(grid.best_params_)
```

```python
#create a random forest model
param_grid = {'n_estimators': [10,50,100], 'max_features': ['auto', 'sqrt']}
model = RandomForestClassifier()
grid = GridSearchCV(model, param_grid, scoring='accuracy')
grid.fit(X_train, Y_train)
Y_pred = grid.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
print(confusion_matrix(Y_test, Y_pred))
print(grid.best_params_)
```

```python
#create a gradient boosting model
param_grid = {'n_estimators': [10,50,100], 'learning_rate': [0.01,0.1,1], 'max_depth': np.arange(3, 10)}
model = GradientBoostingClassifier()
grid = GridSearchCV(model, param_grid, scoring='accuracy')
grid.fit(X_train, Y_train)
Y_pred = grid.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
print(confusion_matrix(Y_test, Y_pred))
```
